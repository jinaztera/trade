```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,y)
plt.show()
```

```python
fig = plt.figure(figsize=(12,6))
plt.plot(x, y)
plt.title("my first line")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
fig.savefig("my_first_pigure.png")
```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, x)
plt.show()
plt.plot(x, np.sin(x), '--')
plt.plot(x, np.cos(x), '.')
plt.show()
plt.plot(x, np.sin(x-5), color='red')
```

```python
random_generator = np.random.RandomState(0)

x = random_generator.randn(50)
y = random_generator.randn(50)

plt.figure(figsize = (9,3))
plt.scatter(x, y)
plt.show()

colors = random_generator.randn(50)
sizes = 1000* random_generator.randn(50)
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)
```
