```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

#   ( )
# df = pd.read_csv('your_ohlcv_data.csv')
#    .
np.random.seed(0)
df = pd.DataFrame(np.random.randn(1000, 5), columns=['Open', 'High', 'Low', 'Volume', 'Close'])

#  
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(df)

#     
def create_dataset(data, time_step=1):
    X, Y = [], []
    for i in range(len(data)-time_step-1):
        a = data[i:(i+time_step), 0:-1]   #      
        X.append(a)
        Y.append(data[i + time_step, -1]) #   (Close)  
    return np.array(X), np.array(Y)

time_step = 100
X, Y = create_dataset(scaled_data, time_step)

#  
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

#  
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

#  
model.fit(X_train, Y_train, batch_size=64, epochs=10, validation_data=(X_test, Y_test))

#   
train_predict = model.predict(X_train)
test_predict = model.predict(X_test)

#   
train_predict = scaler.inverse_transform(np.column_stack((np.zeros((train_predict.shape[0], 4)), train_predict)))
Y_train = scaler.inverse_transform(np.column_stack((np.zeros((Y_train.shape[0], 4)), Y_train.reshape(-1, 1))))
test_predict = scaler.inverse_transform(np.column_stack((np.zeros((test_predict.shape[0], 4)), test_predict)))
Y_test = scaler.inverse_transform(np.column_stack((np.zeros((Y_test.shape[0], 4)), Y_test.reshape(-1, 1))))

#  
train_rmse = np.sqrt(np.mean((Y_train[:, -1] - train_predict[:, -1])**2))
test_rmse = np.sqrt(np.mean((Y_test[:, -1] - test_predict[:, -1])**2))
print(f'Train RMSE: {train_rmse}, Test RMSE: {test_rmse}')
```
